```python
def train_validation_test_split(data, train_size, validation_size):
    train = data[:train_size]
    validation = data[train_size:(train_size+validation_size)]
    test = data[(train_size+validation_size):]
    return train, validation, test
```

```python
train_set, validation_set, test_set = train_validation_test_split(samples, train_size_samples, validation_size_samples)
train_label, validation_label, test_label = train_validation_test_split(labels, train_size_labels, validation_size_labels)
```
