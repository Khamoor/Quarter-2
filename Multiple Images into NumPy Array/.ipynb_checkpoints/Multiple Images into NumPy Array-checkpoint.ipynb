```python
from PIL import Image
import os

imagepath = 'D:\PIAIC Classes\Quarter 2\Multiple Images into NumPy Array\pics'  # Path of file from which image has to be scale
os.chdir(imagepath)
for dirpath, dirnames, filenames in os.walk(os.getcwd()):
    if dirpath == imagepath:
        for f in filenames:
            if f.endswith('.jpg'):
                i = Image.open(f)
                fn, fext = os.path.splitext(f)
                imscaled = i.resize((200, 200))
                imscaled.save('image_200/{}_200{}'.format(fn, fext))
```

```python
import numpy as np
import matplotlib.pylab as plt
import os

imageArray = np.zeros([20,200,200,3])
i = 0
imagepath = 'D:\PIAIC Classes\Quarter 2\Multiple Images into NumPy Array\pics\image_200'
for dirpath, dirnames, filenames in os.walk(os.getcwd()): # Stepping through current directory to destination directory
    if dirpath == imagepath:  # Checking if we get to the destination directory or not
        os.chdir(imagepath)   # Changing directory of current path to image path directory where images are stored
        for f in filenames:   # Iterating over files present in images directory
            if f.endswith('.jpg'): 
                im = plt.imread(f) # Converting image into an numpy array 
                imageArray[i] = im  # Storing images one by one in an defined array
                i += 1
print("ma:\n",imageArray)
print(imageArray.shape)
```
